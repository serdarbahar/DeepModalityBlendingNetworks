```python
import numpy as np 
import matplotlib.pyplot as plt 

y1 = np.zeros((10, 200, 1))
y2 = np.zeros((10, 200, 1))

amplitudes = np.linspace(0.1, 1, 10)

for i in range(10):
    y1[i, :, 0] = amplitudes[i] * np.sin(np.linspace(0, 2 * np.pi, 200))
    y2[i, :, 0] = amplitudes[i] * np.cos(np.linspace(0, 2 * np.pi, 200))

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
for i in range(10):
    plt.plot(y1[i, :, 0], label=f'y1 amplitude {amplitudes[i]:.2f}', color='blue')
plt.title('y1 Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.subplot(1, 2, 2)
for i in range(10):
    plt.plot(y2[i, :, 0], label='y2', color='orange')
plt.title('y2 Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.tight_layout()
plt.show()

np.save('modality1.npy', y1)
np.save('modality2.npy', y2)
```
